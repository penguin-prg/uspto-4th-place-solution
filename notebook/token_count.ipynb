```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
token_counts = json.load(open("/kaggle/input/token-counts/token_counts.json"))
```

```python
counts = []
for v in token_counts.values():
    counts += list(v.values())
counts = np.array(counts)
counts = counts[counts < 1000]
```

```python
len(counts)
```

```python
plt.hist(counts, bins=100, range=(100, 1000))
plt.show()
```
