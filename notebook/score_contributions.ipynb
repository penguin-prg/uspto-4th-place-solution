```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
contributions = []
for i in range(50):
    score = 0
    for j in range(i, 50):
        score += 1 / (j + 1)
    contributions.append(score / 50)
assert np.isclose(sum(contributions), 1)
plt.plot(contributions, marker="o")
plt.grid()
plt.xlabel("i")
plt.ylabel("Contribution")
plt.show()
```

```python
# 積み上げグラフ
plt.bar(range(50), np.array(contributions).cumsum())
plt.grid()
plt.xlabel("i")
plt.ylabel("Contribution")
plt.show()
```
